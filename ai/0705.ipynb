```python
from tensorflow.keras.datasets import fashion_mnist
```

```python
(train_x, train_y), (test_x, test_y) = fashion_mnist.load_data()
```

```python
train_x.shape, train_y.shape
```

```python
test_x.shape, test_y.shape
```

```python
train_y[:10]
```

```python
#convolution2D에서는 입력데이터가 3D이어야 하므로 reshape 해주어야함
train_x = train_x.reshape((60000,28,28,1))
#(-1, 28, 28, 1)도 됨. -1은 크기 모를때
test_x = test_x.reshape((-1,28,28,1))
```

```python
train_x.shape, test_x.shape
```

```python
from tensorflow.keras import models, layers
```

```python
model = models.Sequential()
```

```python
#1st Conv2d - filter_cnt : 32 kernel_size : (3,3) relu
#1st max pooling pool_size=(2,2) 
#2nd Conv2d - filter_cnt : 64 kernel_size : (3,3) relu
#2nd max pooling pool_size=(2,2) 
#3rd Conv2d - filter_cnt : 64 kernel_size : (3,3) relu

model.add(layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu', input_shape=(28,28,1), name='block_1_conv'))
model.add(layers.MaxPool2D(pool_size=(2,2), name='block_1_pool'))

model.add(layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu', name='block_2_conv'))
model.add(layers.MaxPool2D(pool_size=(2,2), name='block_2_pool'))

model.add(layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu', name='block_3_conv'))

#flatten
model.add(layers.Flatten())

model.add(layers.Dense(units=64, activation='relu'))
model.add(layers.Dense(units=10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
```

```python
model.fit(x=train_x, y=train_y, epochs=50, batch_size=128, validation_split=0.2)
```

```python
loss, accuracy = model.evaluate(x=test_x, y=test_y)
```
